### Mis spelled words table
#### We set the cutoff value as 0.8,and if the index of similarity is greater than 0.8, it should be concluded into mis-spelled table

```python
from difflib import SequenceMatcher as sm
```

```python
misspell = []
for x in tag:
    mis = []
    for i in range(len(x)):   
        if x[i] == 'a':
            if sm(a=x,b=x[:i]+'e'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'e' + x[i+1:])
        if x[i] == 'e':
            if sm(a=x,b=x[:i]+'a'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'a' + x[i+1:])
        if x[i] == 'c':
            if sm(a=x,b=x[:i]+'k'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'k' + x[i+1:])
        if x[i] == 'k':
            if sm(a=x,b=x[:i]+'c'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'c' + x[i+1:])
        if x[i] == 'e':
            if sm(a=x,b=x[:i]+'i'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'i' + x[i+1:])
        if x[i] == 'i':
            if sm(a=x,b=x[:i]+'e'+x[i+1:]).ratio() > 0.8:
                mis.append(x[:i] + 'e' + x[i+1:])

        if i <= len(x) - 2:
            if x[i+1] == x[i]:
                if sm(a=x,b=x[0:i]+x[i+1:]).ratio() > 0.8:
                    mis.append(x[0:i]+x[i+1:])

        if i == len(x) - 1 and x[i] == 'e':
            if sm(a=x,b=x+'r').ratio() > 0.8:
                mis.append(x+"r")
        if len(mis) >= 5:
            break
    misspell.append(mis)
```

```python
mis_spelled_table = pd.DataFrame(misspell)
mis_spelled_table = pd.concat( [tagdf, mis_spelled_table], axis=1 )
mis_spelled_table.columns = ['tags','First mis spelled', 'Second mis spelled','Third mis spelled','Fourth mis spelled','Fifth mis spelled','Sixth mis spelled']
mis_spelled_table
```

```python
def process(l):
    delnum = []
    for i in range(len(l)):
        if l[i] == '':
            delnum.append(i)
            continue
        if l[i] in l[:i]:
            delnum.append(i)
            continue
    for i in reversed(delnum):
        l.pop(i)
    return l 
```
