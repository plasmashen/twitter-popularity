```python
def process(l):
    delnum = []
    for i in range(len(l)):
        if l[i] == '':
            delnum.append(i)
            continue
        if l[i] in l[:i]:
            delnum.append(i)
            continue
    for i in reversed(delnum):
        l.pop(i)
    return l 
```

```python
import datetime
import time
def get_sec(time_str):
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
result=mdb.tweettable.find({'followers':{'$gt':1000}},{'created_at':1})
df=pd.DataFrame(([i['created_at'] for i in result]))
j=0
for i in df[0]:
    x=datetime.datetime.strftime(i,'%H:%M:%S')
    j=j+(get_sec(x))
str(datetime.timedelta(seconds=j/len(df)))
```
